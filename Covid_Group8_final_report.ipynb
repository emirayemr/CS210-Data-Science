### kNN Method

```python
# taken from Hypothesis_Testing.ipynb (recit 6)
def create_2D_sample(mean, std, n, label):
  """
  given a set of sample parameters
  returns a 2D sample
  """
  # create data points
  xvals = np.random.normal(loc=mean, scale=std, size=n).reshape(-1,1)
  yvals = np.random.normal(loc=mean, scale=std, size=n).reshape(-1,1)

  
  # class labels
  labels = np.repeat(label, n).reshape(-1,1)

  # merge observations and assigned labels
  sample = np.hstack([xvals, yvals, labels])

  return sample
```

```python
n = 20

means = [africa_conf_mean, europe_conf_mean, eastmed_conf_mean]
stds = [africa_conf_std, europe_conf_std, eastmed_conf_std]
colors = ["green", "blue", "yellow"]

np.random.seed(42)
d = []
i = 0
while i < 3:
  sample = create_2D_sample(means[i], stds[i], n, i)
  d.append(sample)
  i += 1

data = np.vstack(d)

plt.scatter(data[:,0], data[:,1], c=data[:,2], s=20)
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

In this figure we observe randomly generated data from each distinct classes, classes are from three continent each point has a distinct coordinate. In the following figures prediction will be made with respect to observed data.

```python
point = np.array([[0,0]])

plt.scatter(data[:,0], data[:,1], c=data[:,2], s=20)
plt.scatter(point[:,0], point[:,1], s=80, c="red")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

In this figure there is a new data point which is red we don't know since it's not labeled we don't know it's class so we can choose small k or large k in order to label this new dot to any class we  will use euclodian method to calculate distances between a new dot and other different colored data point hence we will determine which class or which continent this data point belongs to.

```python
from sklearn.neighbors import KNeighborsClassifier

k = 3
model = KNeighborsClassifier(k, metric="euclidean")

features = data[:, :2]
labels = data[:, 2]

model.fit(features, labels)

pred = model.predict(point)

plt.scatter(data[:,0], data[:,1], c=data[:,2], s=20)
plt.scatter(point[:,0], point[:,1], s=80, c=pred)
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

By using euclodian distance we have predicted that this new data point belongs to European continent

In this figure by the red we want to estimate the confirmed cases in the world in time intervals since the black dots are the observed confirmed cases we would like to use linear regression method which will help us to predict confirmed cases in the world

```python
# taken from Regression.ipynb (recit 10)
def r2(y_true, y_pred):
  u = sum((y_true - y_pred)**2)
  v = sum((y_true - y_true.mean())**2)

  return 1-(u/v)

r2_coeff = r2(y, reg_line)
print(r2_coeff)
```

### Prediction of Death Counts for 7 Days

```python
def plot_reg_R2(days, m_reg, b_reg):
  m = 600000
  b = 150

  # creating random samples
  np.random.seed(0)
  x = 10*np.random.rand(days)
  y = m*x + b + np.random.randn(days)

  # regression line
  reg_line = m_reg*x + b_reg

  # plotting the residuals
  for i in range(len(x)):
    plt.plot((x[i], x[i]), (reg_line[i], y[i]), 'b')

  # plotting the regression line
  plt.plot(x, reg_line, c="red")
  plt.scatter(x, y, s=10, marker="s", c="black")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.title(r"$R^2: {:.2f}$".format(r2_coeff))

interact(plot_reg_R2, days=7, m_reg=200000, b_reg=600000)
plt.show()
```

Randomly generated points are above the regression line, so it is predicted   that death counts will increase next 7 days.
